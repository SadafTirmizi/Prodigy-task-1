```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("./train.csv")
df
```

```python
df.fillna(0,inplace=True)
df
```

```python

X_train = df[['LotArea', 'PoolArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'GarageArea', 'OpenPorchSF', 'EnclosedPorch', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr']].values

y_train = df['SalePrice'].values.reshape(-1, 1)  

model = LinearRegression()

model.fit(X_train, y_train)
y_pred = model.predict(X_train)


print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
